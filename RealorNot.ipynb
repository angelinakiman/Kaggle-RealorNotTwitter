<h2>Data Preparation</h2>
<h3>Cleaning Data</h3>

<h4>Cleaning text used for text-mining</h4>

```python
def clean_text(data, text):
    data[text] = data[text].str.replace(r"http\S+", "") # remove URLs
    data[text] = data[text].str.replace(r"http", "") # remove mention of http if not URL
    data[text] = data[text].str.replace(r"@\S+", "") # remove any alias
    data[text] = data[text].str.replace(r"[^A-Za-z0-9(),!?@\'\`\"\_\n]", " ") # remove any character not in this list
    data[text] = data[text].str.replace(r"@", "at") # replace non-alias @ with at
    data[text] = data[text].str.lower() # make everything lowercase
    return data

data = clean_text(data, "text")

data.head(5) # show first 5 lines
```

<h4>Tokenize text by separating it into individual words</h4>

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')

data["tokens"] = data["text"].apply(tokenizer.tokenize)
data.head(5) # show first 5 lines
```
